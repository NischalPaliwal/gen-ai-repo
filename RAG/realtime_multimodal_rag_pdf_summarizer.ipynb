````python
from unstructured.partition.pdf import partition_pdf
````

````python
elements = partition_pdf(
    filename="/content/data/NPTEL Week 10 Exam Notes.pdf",          # Mandatory parameter
    strategy="hi_res",                                              # Mandatory parameter
    extract_images_in_pdf=True,                                     # Mandatory parameter
    extract_image_block_types=["Image", "Table"],
    extract_image_block_to_payload=False,
    extract_image_block_output_dir="extracted_data"
)
````

```markdown
Let's break down the parameters used in the `partition_pdf` function:

*   `filename`: This is a mandatory parameter that specifies the path to the PDF file you want to process. In this case, it's `/content/data/INM - Week - 6.pdf`.

*   `strategy`: This is also a mandatory parameter that dictates how the PDF content is extracted. Setting it to `"hi_res"` (high-resolution) typically means the function will use more advanced techniques, often involving OCR (Optical Character Recognition) and layout analysis, to get a more accurate and detailed extraction of text and elements.

*   `extract_images_in_pdf`: When set to `True`, this boolean parameter instructs the function to identify and extract images embedded within the PDF document. These images are often crucial for a complete understanding of the document's content.

*   `extract_image_block_types`: This parameter specifies the types of image blocks you want to extract. Here, `["Image", "Table"]` indicates that the function should look for both standalone images and images that are part of tables. This is particularly useful for documents where tables might be represented as images.

*   `extract_image_block_to_payload`: If set to `True`, the extracted image blocks (specified by `extract_image_block_types`) would be embedded directly into the element objects returned by `partition_pdf`. However, with `False`, the images are not included in the payload, which can be useful if you only need the text or if the images are very large.

*   `extract_image_block_output_dir`: This parameter specifies a directory where the extracted image files will be saved. In this example, `"extracted_data"` means all extracted images will be stored in a folder named `extracted_data` relative to the current working directory.

In summary, this configuration aims to perform a thorough, high-resolution extraction of both text and image-based content from the specified PDF, saving the extracted images to a local directory.
```

````python
prompt_template = ChatPromptTemplate.from_template(prompt_text)
````

## Image Summary

````python
import base64
import os
from langchain.messages import HumanMessage
````

````python
def encode_image(image_path):
  with open(image_path, "rb") as file:
    return base64.b64encode(file.read()).decode("utf-8")
````

````python
def image_summarize(base64_string, prompt):
  llm = ChatOllama(
      model="llama3.2-vision",
      temperature=1.0
  )

  response = llm.invoke(
      [
          HumanMessage(
                content=[
                    {"type": "text", "text": prompt},
                    {
                        "type": "image_url",
                        "image_url": {"url": f"data:image/jpeg;base64,{base64_string}"},
                    },
                ]
            )
      ]
  )

  return response.content
````

````python
def generate_image_summaries(path):
    base64_string_list = []
    image_summaries = []

    prompt = """You are an assistant tasked with summarizing images for retrieval. \
    These summaries will be embedded and used to retrieve the raw image. \
    Give a concise summary of the image that is well optimized for retrieval."""

    for image_file in sorted(os.listdir(path)):
        if image_file.endswith(".jpg"):
            image_path = os.path.join(path, image_file)
            base64_image = encode_image(image_path)
            base64_string_list.append(base64_image)
            image_summaries.append(image_summarize(base64_image, prompt))


    return base64_string_list, image_summaries
````

````python
base64_string_list, image_summaries = generate_image_summaries(images_path)
````
